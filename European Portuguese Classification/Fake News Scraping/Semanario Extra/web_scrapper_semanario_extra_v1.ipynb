```python
from bs4 import BeautifulSoup as bs4
import requests
import pandas as pd
import re
import time

pages=[]
titles=[]
texts=[]
sources=[]
urls=[]

# Initialize an empty dataframe
df = pd.DataFrame(columns=['Title', 'Text', 'Source', 'URL'])
```

```python
for url in pages:
    page = requests.get(url)
    soup = bs4(page.text, 'html.parser')

    link_elements = soup.find_all("div", class_="td-module-thumb")

    for link_element in link_elements:
        link = link_element.find("a")
        if link.get("href") not in urls:
            urls.append(link.get("href"))
            print(link.get("href"))

    time.sleep(3)

# Print the extracted links
# for url in urls:
#     print(url)
```

```python
len(urls)
```

```python
for url in urls:
    page = requests.get(url)
    soup = bs4(page.text, 'html.parser')

    news_title = soup.find("div", class_="td-post-header").get_text()
    news_content = soup.find("div", class_="td-post-content").get_text()

    # Remove first blank paragraph in title
    news_title = news_title.split('\n', 1)[1]

    # Remove extra blank lines between paragraphs
    news_title = re.sub(r'\n{3,}', '\n\n', news_title)
    news_content = re.sub(r'\n{3,}', '\n\n', news_content)

    # Try finding source
    try:
        news_source = soup.find("div", class_="td-post-small-box").get_text().split('FONTE')[-1]
    except:
        news_source = "n/a"

    titles.append(news_title)
    texts.append(news_content)
    sources.append(news_source)

    time.sleep(3)

# Create a new dataframe from the lists
new_data = pd.DataFrame({'Title': titles, 'Text': texts, 'Source': sources, 'URL': urls}) #, 'URL': unique_urls_list

# Concatenate the new dataframe with the existing dataframe
df = pd.concat([df, new_data], ignore_index=True)
```
