```python
from bs4 import BeautifulSoup as bs4
import requests
import pandas as pd
import re
import time

pages=[]

titles=[]
texts=[]
urls= set()
category_pages=[]

# Initialize an empty dataframe
df = pd.DataFrame(columns=['Title', 'Text', 'URL'])
```

```python
main_url = "https://www.tuga.press/"
pages.append(main_url)
for i in range(2,129):
    pages.append(main_url+'page/'+str(i))

for page in pages:
    print(page)
```

```python
for url in pages:
    print(url)
    print('--------------------------')
    page = requests.get(url)
    soup = bs4(page.text, 'html.parser')
    
    # Extract articles' links
    try:
        link_elements = soup.find_all("h2")
        for link_element in link_elements:
            links = link_element.findAll('a')
            for link in links:
                extra_link = link.get('href')
                #print('a',extra_link)
                urls.add(extra_link)
    except:
        pass

    # Get recommended articles' link
    try:
        link_elements = soup.find_all("div", class_ = 'dpArticleShare_MostSharedEents')
        for link_element in link_elements:
            links = link_element.findAll('li')
            for link in links:
                a_element = link.findAll('a')
                for recommended_article_link in a_element:
                    print(recommended_article_link.get('href'))
                    urls.add(recommended_article_link.get('href'))
    except:
        pass

    time.sleep(3)
```

```python
final_urls = set(urls)  # Track processed URLs to avoid duplicates

# Cycle through urls while dynamically adding recommended urls to both urls list and final_urls set
while urls:
    print('set_len:',len(urls))
    url = urls.pop()  # Get the first URL from the list
    print("webpage:",url)
    page = requests.get(url)
    soup = bs4(page.text, 'html.parser')
        
    # Get recommended articles' link
    try:
        link_elements = soup.find_all("div", class_ = 'dpArticleShare_MostSharedEents')
        for link_element in link_elements:
            links = link_element.findAll('li')
            for link in links:
                a_element = link.findAll('a')
                for recommended_article_link in a_element:
                    if recommended_article_link.get('href') not in final_urls:
                        print(recommended_article_link.get('href'))
                        urls.add(recommended_article_link.get('href'))
                        final_urls.add(recommended_article_link.get('href'))
    except:
        pass
                    
    time.sleep(3)
```

```python
titles=[]
texts=[]
sources=[]
df_urls = []

# Initialize an empty dataframe
df = pd.DataFrame(columns=['Title', 'Text', 'Source', 'URL'])

for url in final_urls:
    print(url)
    page = requests.get(url)
    soup = bs4(page.text, 'html.parser')
    try:
        news_title = 'NONE'
        news_text = 'NONE'
        news_source = 'NONE'
        news_category = 'NONE'
        news_tags = []

        news_title = soup.find("h1").get_text().strip()
        paragraphs = soup.find_all("p")
        #print(paragraphs)
        text = "\n".join(p.get_text() for p in paragraphs)
        lowercase_text = text.lower()
        if 'fonte:' in lowercase_text:
            index = lowercase_text.index('fonte:')
            news_text = text[:index]
            news_source = text[index + len('fonte:'):].split('\n')[0].strip()
        else:
            news_text = text
            news_source = 'n/a'
        
        news_text = news_text.strip()
        lines = news_text.splitlines()

        # Remove tuga.press
        stripped_line = lines[0].lower().strip()
        if ("tuga.press" in stripped_line):
            lines = lines[1:]
        
        # Remove V aqui notcias parecidas com esta que acabaste de ler
        stripped_line = lines[-1].strip()
        if "V aqui notcias parecidas com esta que acabaste de ler:" in stripped_line:
            lines = lines[:-1]

        # Join the lines back into a string
        news_text = '\n'.join(lines).lstrip()
        news_text = re.sub(r'\n+', '\n', news_text) # Replace consecutive paragraphs with a single one
        news_text = re.sub(r'\t', '', news_text) # Replace consecutive paragraphs with a single one
    except:
        pass

    print(news_title)
    print(news_text)
    print(news_source)

    titles.append(news_title)
    texts.append(news_text)
    sources.append(news_source)
    df_urls.append(url)

    time.sleep(3)

# Create a new dataframe from the lists
new_data = pd.DataFrame({'Title': titles, 'Text': texts, 'Source': sources, 'URL': df_urls})

# Concatenate the new dataframe with the existing dataframe
df = pd.concat([df, new_data], ignore_index=True)
```
