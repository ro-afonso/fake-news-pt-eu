```python
for url in pages:
    print("next url:",url)
    page = requests.get(url)
    soup = bs4(page.text, 'html.parser')
    #print(soup.prettify())
    # filter desired urls
    main_url = url.split("/https")[0]
    for link in soup.findAll('a'):
        if (link.get("href") != main_url+'/https://bombeiros24.pt/contacto/' and 
            link.get("href") != main_url+'/https://bombeiros24.pt' and
            link.get("href") != main_url+'/https://bombeiros24.pt/' and
            link.get("href") != main_url+'/https://bombeiros24.pt/category/noticias/' and
            link.get("href") != main_url+'/https://bombeiros24.pt/category/critica/' and
            link.get("href") != main_url+'/https://bombeiros24.pt/category/meteorologia/' and
            link.get("href") != main_url+'/https://bombeiros24.pt/category/dicas/' and
            link.get("href") != main_url+'/https://bombeiros24.pt/category/uncategorized/' and
            link.get("href") != main_url+'/https://bombeiros24.pt/category/outros/' and
            link.get("href") != main_url+'/https://bombeiros24.pt/author/admin/' and
            link.get("href") != main_url+'/https://www.facebook.com/osnossosbombeiros/' and
            link.get("href") != main_url+'/https://bombeiros24.pt/politica-de-privacidade/' and
            link.get("href") != main_url+'/https://bombeiros24.pt/page/2/' and
            link.get("href") != main_url+'/https://bombeiros24.pt/page/3/' and
            link.get("href") != main_url+'/https://bombeiros24.pt/page/4/' and         
            link.get("href") != '#' and link.string != 'Página seguinte »' and link.get("href") is not None):
            urls.append(link.get("href"))
            print(link.get("href"))
    time.sleep(3)
```

```python
unique_urls = set(urls)

for link in unique_urls:
    print(link)
    
print(len(unique_urls))
```

```python
unique_urls_list = list(unique_urls)
urls.clear()
titles.clear()
texts.clear()
df = pd.DataFrame(columns=['Title', 'Text', 'URL'])
#print(unique_urls_list[0])

for url in unique_urls_list:
    print(url)
    page = requests.get(url)
    soup = bs4(page.text, 'html.parser')
    # get title and texts
    for item in soup.findAll('h1'):
        try:
            title = item.getText()
            if title not in titles:
                # only append unique entries

                # content in div class= entry themeform
                div_text=soup.find("div",{"class":"entry themeform"}).get_text()
                #print(div_text)

                # Remove extra blank lines between paragraphs
                div_text = re.sub(r'\n{3,}', '\n\n', div_text)

                # Remove the first 5 paragraphs
                for i in range(1,5):
                    div_text = div_text.split('\n', 1)[1]

                # Remove the last 6 paragraphs
                lines = div_text.split('\n')
                text = '\n'.join(lines[:-6])

                titles.append(title)
                texts.append(text)
                urls.append(url)
        except:
            print(url,"ignored")
        
    time.sleep(3)

# Create a new dataframe from the lists
new_data = pd.DataFrame({'Title': titles, 'Text': texts, 'URL': urls}) #, 'URL': unique_urls_list

# Concatenate the new dataframe with the existing dataframe
df = pd.concat([df, new_data], ignore_index=True)
```
