```python
import pandas as pd
```

```python
df_2017 = pd.read_csv('Bombeiros24_dataset_2017.csv')
df_2018 = pd.read_csv('Bombeiros24_dataset_2018.csv')
df_2019_p1 = pd.read_csv('Bombeiros24_dataset_2019_p1.csv')
df_2019_p2 = pd.read_csv('Bombeiros24_dataset_2019_p2.csv')
df_2020 = pd.read_csv('Bombeiros24_dataset_2020.csv')
```

```python
merged_df = pd.concat([df_2017, df_2018, df_2019_p1, df_2019_p2, df_2020], ignore_index=True)
```

```python
merged_df.to_csv('bombeiros24_merged_dataset.csv', index=False)
```

```python
final_df = pd.read_csv("bombeiros24_merged_dataset.csv")
```

```python
# Some articles include reputable sources at the end, which indicates the possibility of not being fake

# Set the maximum width to display the entire text
pd.set_option('display.max_colwidth', None)

# Create a boolean mask to check if the text ends with well-known sources
mask_RR = final_df['Text'].str.strip().str.endswith('RR', na=False)
mask_DN = final_df['Text'].str.strip().str.endswith('DN', na=False)
mask_DN_full = final_df['Text'].str.strip().str.endswith('Diário de Noticias', na=False)
mask_JN = final_df['Text'].str.strip().str.endswith('JN', na=False)
mask_JN_full = final_df['Text'].str.strip().str.endswith('Jornal de Noticias', na=False)
mask_Observador = final_df['Text'].str.strip().str.endswith('Observador', na=False)
mask_TVI_24 = final_df['Text'].str.strip().str.endswith('TVI 24', na=False)
mask_TVI24 = final_df['Text'].str.strip().str.endswith('TVI24', na=False)
```

```python
# Create a combined mask to identify rows to be removed
combined_mask = mask_DN | mask_DN_full | mask_JN | mask_JN_full | mask_Observador | mask_TVI_24 | mask_TVI24

# Remove the rows that match the combined mask
final_df_without_sources = final_df.drop(final_df[combined_mask].index)
```

```python
# Check the last line of text of each row
for index, row in final_df_without_sources.iterrows():
    # Split the text into lines and access the last line
    lines = row['Text'].splitlines()
    last_line = lines[-1]
        
    # Print the last line of text
    print(last_line)
```

```python
# Iterate over each row in the DataFrame
for index, row in final_df_without_sources.iterrows():
    # Split the text into lines and access the last line
    lines = row['Text'].splitlines()
    last_line = lines[-1]

    # Check if the last line contains 'Fonte' and drop it if so
    if 'Fonte' in last_line:
        final_df_without_sources.drop(index, inplace=True)
```

```python
# Check final dataset
for index, row in final_df_without_sources.iterrows():
    # Split the text into lines and access the last line
    lines = row['Text'].splitlines()
    last_line = lines[-1]
        
    # Print the last line of text
    print(last_line)
```
