Direita Politica

```python
direita_politica_df = pd.read_csv("Direita Politica\Direita_politica_dataset.csv")
direita_politica_df.head()
```

```python
len(direita_politica_df)
```

```python
# Check for duplicate rows
direita_politica_df.duplicated().sum()
```

```python
# Delete duplicate rows
direita_politica_df.drop_duplicates(inplace=True)
```

```python
len(direita_politica_df)
```

```python
unique_sources = set(direita_politica_df['Source'])
for source in unique_sources:
    print(source)

print(len(unique_sources))
```

```python
filtered_rows = []
for index, row in direita_politica_df.iterrows():
    source_df = row["Source"]
    if pd.isnull(source_df):
        filtered_rows.append(row)
    else:
        source_matched = False
        for source in credible_sources_list:
            if str(source).lower() in str(source_df).lower():
                source_matched = True
                break
        if not source_matched:
            filtered_rows.append(row)

direita_politica_filtered_df = pd.DataFrame(filtered_rows)
```

```python
unique_sources = set(direita_politica_filtered_df['Source'])
for source in unique_sources:
    print(source)

print(len(unique_sources))
```

```python
len(direita_politica_filtered_df)
```

```python
# Merge title and text into a single column
direita_politica_filtered_df['Text'] = direita_politica_filtered_df['Title'] + ' ' + direita_politica_filtered_df['Text']
direita_politica_filtered_df.drop(columns=['Title'],inplace=True)

direita_politica_filtered_df.head()
```

Fact-checks label conversion - Publico

```python
publico_fact_check_p1_df = pd.read_csv("Publico Fact Check\Publico_fact_check_dataset_p1.csv")
publico_fact_check_p2_df = pd.read_csv("Publico Fact Check\Publico_fact_check_dataset_p2.csv")

publico_fact_check_merged_df = pd.concat([publico_fact_check_p1_df, publico_fact_check_p2_df], ignore_index=True)
publico_fact_check_merged_df.head()
```

```python
print(set(publico_fact_check_merged_df["Label"]))

# Mapping for label conversion
label_mapping = {
    "Parcialmente verdadeiro": 1,
    "Verdadeiro": 1,
    "Parcialmente falso": 0,
    "Falso": 0,
    "Inconclusivo": 0,
}

# Convert labels using mapping
publico_fact_check_merged_df["Label"] = publico_fact_check_merged_df["Label"].replace(label_mapping)
```

```python
publico_fact_check_merged_df.head()
```

```python
print(set(publico_fact_check_merged_df['Label']))
```

```python
publico_fact_check_merged_df.to_csv("Publico_fact_check_converted.csv", index=False)
```

```python
publico_fact_check_fake_df = publico_fact_check_merged_df[publico_fact_check_merged_df["Label"] == 0].drop("Label", axis=1)

publico_fact_check_fake_df.head()
```

```python
publico_fact_check_fake_df.to_csv("Publico_fact_check_fake.csv", index = False)
```

```python
publico_fact_check_real_df = publico_fact_check_merged_df[publico_fact_check_merged_df["Label"] == 1].drop("Label", axis=1)
publico_fact_check_real_df.head()
```
